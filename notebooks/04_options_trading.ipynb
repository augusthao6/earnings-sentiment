# Options Trading on Earnings Sentiment

This notebook demonstrates the end-to-end pipeline for predicting straddle opportunities from earnings transcripts.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

PROJECT_ROOT = Path('..')
DATA_DIR = PROJECT_ROOT / 'data' / 'processed'
```

## Load Training Data

```python
df = pd.read_csv(DATA_DIR / 'training_data.csv')
df.head()
```

## Model Performance

```python
# Load backtest results
results = pd.read_csv(DATA_DIR / 'backtest_results.csv')

# Plot cumulative P&L
results['cum_pnl'] = results['pnl'].cumsum()
plt.figure(figsize=(10, 6))
plt.plot(results['cum_pnl'])
plt.title('Cumulative P&L')
plt.xlabel('Trade')
plt.ylabel('P&L')
plt.show()

print(f"Total P&L: {results['pnl'].sum():.4f}")
print(f"Win Rate: {(results['pnl'] > 0).mean():.2%}")
```
